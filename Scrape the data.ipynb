# Let us scrape the data from the wikipedia page

https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

As a first step we install the required libraries

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from bs4 import BeautifulSoup

import json # library to handle JSON files

import requests # library to handle requests

print('Libraries imported.')
```

Next we store the url of the wikipedia page from where we wish to retrieve the data and use beautifulsoup to parse it 

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(website_url,'html.parser')
print("Parsed successfully")
```

Now as per the question, we need to form a dataframe of postal codes and neighbourhoods, which are stored in a table, so we parse and view the table in the HTML page

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

Let us now look at the columns present in the table

```python
print(My_table.tr.text)
```

```python
trs = My_table.find_all('tr')
rows = []
for tr in trs:
    i = tr.find_all('td')
    if i:
        rows.append(i)

lst = []
for row in rows:
    postalcode = row[0].text.rstrip()
    borough = row[1].text.rstrip()
    neighborhood = row[2].text.rstrip()
    if borough != 'Not assigned':
        lst.append([postalcode, borough, neighborhood])
        
lst
```

We now proceed to create a dataframe of our data

```python
cols=cols = ['PostalCode', 'Borough', 'Neighbourhood']
df = pd.DataFrame(lst, columns=cols)
print(df.shape)
```

```python
df.head(15)
```

Now as a data processing step, as required, wherever the neighbourhood is not assigned, we use Borough as the Neighbourhood

```python
df.loc[df['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = df['Borough']
df.head(15)
```

A single postal code can have multiple neighbourhoods, we proceed to group the data by postal code and re-index it

```python
result = df.groupby(['PostalCode','Borough'], sort=False).agg( ', '.join)
df_new=result.reset_index()
df_new.head(15)
```
